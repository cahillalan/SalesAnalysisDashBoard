### KPI Analysis - Sales Analytics Dashboard

## Objectives
Creat key business metrics (KPIs) to support dashboard development and decision_making.

## KPI focus
- Revenue performance
- Profitability
- Order behaviour
- Product and regional performance

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/clean_superstore.csv")
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.strftime('%b')
```

```python
total_sales =df['sales'].sum()
total_profit = df['profit'].sum()
total_orders = df['order_id'].nunique()
total_quantity = df['quantity'].sum()

print("Total Sales:", total_sales)
print("Total Profit:" , total_profit)
print("Total Orders:" , total_orders)
print("Total Quantity:", total_quantity)
```

### Core KPIs

These metrics provide a high-level view of business performance:
- Total Sales reflects overall revenue
- Tota Profit indicates profitability
- Total Orders shows transaction volume
- Total Quantity reflects units sold

```python
df['profit_margin'] = df['profit']/df['sales']

avg_profit_margin =df['profit_margin'].mean()

avg_profit_margin
```

### Profit Margin

Profit margin measures how efficiently revenue is converted into profit.

Lower margins may indicate high costs or excessive discounting.

```python
aov = df.groupby('order_id')['sales'].sum().mean()

aov
```

### Average Order Value

AOV represents teh average revenue per transaction.

This helps understand customer purchasing behaviour.

```python
monthly_kpis = df.groupby(pd.Grouper(key='order_date' , freq='M')).agg({
    'sales':'sum',
'profit':'sum',
'order_id':'nunique'
    }).reset_index()

monthly_kpis.rename(columns={'order_id':'orders'},inplace=True)
monthly_kpis.head()
```

### Monthly KPIs

Monthly aggregation enables trend analysis and supports time-based dashboard visuals.


```python
category_kpis = df.groupby('category').agg({
    'sales':'sum',
    'profit':'sum',
    'order_id':'nunique'
}).sort_values(by='sales',ascending =False).reset_index()

category_kpis.rename(columns={'order_id':'orders'},inplace=True)

category_kpis
```

### Catregory Performance

Identifies top-performing categories and their contribution to revenue and profit.

```python
regional_kpis = df.groupby('region').agg({
    'sales':'sum',
    'profit':'sum',
    'order_id':'nunique'
}).sort_values(by='sales',ascending =False).reset_index()

regional_kpis.rename(columns={'order_id':'orders'},inplace=True)

regional_kpis
```

### Regional Performance

Highlights geographic differences in sales and profit

```python
product_kpis = df.groupby('product_name').agg({
    'sales':'sum',
    'profit':'sum'
}).sort_values(by='sales',ascending =False).reset_index()


product_kpis.head(10)
```

### Product Performance

Identifies high-performing and under-performing products.

```python
discount_kpis = df.groupby('discount').agg({
    'sales':'sum',
    'profit':'sum'
}).reset_index()

discount_kpis
```

### Discount Impact

Analysis shows that higher discounts often reducs profitability.

This insight suggests that discount strategies should be carefully controlled.

```python
customer_kpis = df.groupby('customer_id').agg({
    'sales':'sum',
    'order_id':'nunique'
}).rename(columns={'order_id':'orders'})

customer_kpis
```

### Customer Behaviour

Customer-level metrics help identify high-value customers and purchasing patterns.

```python
df.to_csv("../data/processed/dashboard_data.csv", index =False)
monthly_kpis.to_csv("../data/processed/monthly_kpis.csv",index=False)
category_kpis.to_csv("../data/processed/category_kpis.csv",index=False)
regional_kpis.to_csv("../data/processed/regional_kpis.csv",index=False)
product_kpis.to_csv("../data/processed/product_kpis.csv",index=False)
discount_kpis.to_csv("../data/processed/discount_kpis.csv",index=False)
customer_kpis.to_csv("../data/processed/customer_kpis.csv",index=False)
```

```python
kpi_summary = pd.DataFrame({
    'total_sales':[df['sales'].sum()],
    'total_profit':[df['profit'].sum()],
    'total_orders':[df['order_id'].nunique()],
    'avg_order_value':[df.groupby('order_id')['sales'].sum().mean()],
    'avg_profit_margin':[(df['profit']/df['sales']).mean()]
                         })
kpi_summary.to_csv("../data/processed/kpi_summary.csv", index=False)
```
